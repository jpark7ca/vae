### 
### VAE Model Architecture

```python
class View(nn.Module):
    def __init__(self, *args):
        super().__init__()
        self.shape = args

    def forward(self, x):
        return x.view(self.shape)

class Encoder(nn.Module):
    def __init__(self, debug=False):
        super().__init__()
        self.debug = debug
        
        self.encoder = nn.Sequential(
            nn.Conv2d(in_channels=1, out_channels=32, stride=(1, 1), kernel_size=(3, 3), padding=1),
            nn.SiLU(),
            nn.Conv2d(in_channels=32, out_channels=64, stride=(2, 2), kernel_size=(3, 3), padding=1),
            nn.SiLU(),
            nn.Conv2d(in_channels=64, out_channels=64, stride=(2, 2), kernel_size=(3, 3), padding=1),
            nn.SiLU(),
            nn.Flatten(),  # *** flatten except the first dimension which is considered as a batch dimension. ***
        )

    def forward(self, x):
        return self.encoder(x)

class Decoder(nn.Module):
    def __init__(self, debug=False):
        super().__init__()

        self.decoder = nn.Sequential(
            nn.Linear(2, 3136),
            View(-1, 64, 7, 7),
            nn.ConvTranspose2d(64, 64, stride=(2, 2), kernel_size=(4, 4), padding=1), # used for upsampling
            nn.SiLU(),
            nn.ConvTranspose2d(64, 32, stride=(2, 2), kernel_size=(3, 3), padding=1),
            nn.SiLU(),
            nn.ConvTranspose2d(32, 1, stride=(1, 1), kernel_size=(4, 4), padding=1),
            nn.Tanh()
        )

    def forward(self, x):
        return self.decoder(x)
        
class VariationalAutoEncoder(nn.Module):
    def __init__(self, debug=False):
        super().__init__()
        
        self.encoder = Encoder(debug)
        self.decoder = Decoder(debug)

        self.z_mu = nn.Linear(3136, 2)
        self.z_logvar = nn.Linear(3136, 2) # learn logvar instead of var *** logvar trick

    def reparameterize(self, mu, var):
        # add noise (randomness)
        err = torch.randn(mu.shape).to(mu.get_device())
        
        # log var trick
        z = mu + err*torch.exp(var/2.0)
        
        return z

    def forward(self, x):
        encoded = self.encoder(x)
        mu = self.z_mu(encoded)
        logvar = self.z_logvar(encoded)
        z = self.reparameterize(mu, logvar)
        decoded = self.decoder(z)

        return mu, logvar, decoded
        
```

### 
### Training Loop

```python
# Instantiate a VAE model
vae = VariationalAutoEncoder(debug=False)
vae.to(DEVICE)

# optimizier
optimizer = torch.optim.Adam(vae.parameters(), lr=LEARNING_RATE)

# scheduler
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=STEP_SIZE, gamma=GAMMA)

# evaluation
losses = []
x_axis = []
y_axis = []


# training loop
for epoch in range(1, EPOCHS+1):
    # train
    vae.train(True)

    _losses = []
    for index, (image, label) in enumerate(tqdm(train_loader)):
        # image[0]: image
        # image[1]: label
        actual = image.to(DEVICE)
        z_mu, z_logvar, pred = vae(actual)
        
        batch_size = image.shape[0]
        # MSE (L1)
        mse = F.mse_loss(input=pred, target=actual, reduction='sum')/batch_size
        # KL Divergence (L2)
        kl = (-0.5 * torch.sum(1 + z_logvar - z_mu**2 - torch.exp(z_logvar), axis=1)).mean()
        # Loss =  MSE (L1) + KL (L2)
        loss = mse + kl

        optimizer.zero_grad(True)
        
        loss.backward()

        # update model parameters
        optimizer.step()

        _losses.append(loss.item())
        losses.append(loss.item())
    
    scheduler.step()
    
    print(f"epoch: {epoch}, scheduler-step: {scheduler.last_epoch}, batch loss: {np.array(_losses).mean():.4f}, learning rate: {optimizer.param_groups[0]['lr']:.6f}")
    
    x_axis.append(epoch)
    y_axis.append(np.array(losses).mean())

    print(f"epoch: {epoch}, avg. loss: {np.array(losses).mean()}")
    
```

###
### Evaluation of Model Training

```python
plt.plot(x_axis, y_axis)
plt.title('Learning Curves')
plt.xlabel('epoch')
plt.ylabel('error')
plt.legend(['Training'])
plt.show()
```
